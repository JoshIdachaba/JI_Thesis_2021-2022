**HTML Parser for NHANES Data Dictionary Information**

With the Diet & Behavior Quality section as a test case, extract the question information from the 

```python
from bs4 import BeautifulSoup
import urllib.request
import re

```

```python
url = urllib.request.urlopen("https://wwwn.cdc.gov/Nchs/Nhanes/2017-2018/DBQ_J.htm")

nhanes_dbq = url.read()

nhanes_dbq_html= BeautifulSoup(nhanes_dbq, "html.parser")

```

*If needed, write HTML to file*

```python
filename = input("Save as, including '.html':")

with open(filename, "w", encoding="utf-8") as file:
    file.write(str(nhanes_dbq_html))

```

*If desired or needed, open html file*

```python
with open("nhanes_dbq.html") as file:
    nhanes_dbq = file.read()

nhanes_dbq_html= BeautifulSoup(nhanes_dbq, "html.parser")
```

***Parse NHANES Data***

```python
body = nhanes_dbq_html.body

print(body)
```

```python
codebook = body.find_all("div", id = "codebook")
print(codebook)
```

```python
def element_length(html,index, tag, el_class):
    return len(str(html.find_all("dd", class_ = "info")[index]))

print(len(str(body.find_all("dd", class_ = "info")[2])))

print(element_length(body,2,"dd","info"))
```

```python
DBQ_titles = body.find_all("dd", class_ = "info", text = re.compile("DBQ*"))
for i in DBQ_titles:
    print(i)
```
