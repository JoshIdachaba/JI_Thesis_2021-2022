```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

```python
ingredient_lists = {}
for index,row in OpenFoodFacts_Demo.iterrows():
    ingredient_lists[row['code']] = row['ingredients_text']
    
documents = list(ingredient_lists.values()) 
print(documents)
```

```python
OpenFoodFacts_Demo = pd.read_csv("OpenFoodFacts_Head_Demo.csv")
OpenFoodFacts_Demo.fillna({'product_name':"Unknown"}, inplace=True)
```

```python
count_vectorizer = CountVectorizer(stop_words='english')
count_vectorizer = CountVectorizer()
sparse_matrix = count_vectorizer.fit_transform(documents)
```

```python
doc_term_matrix = sparse_matrix.todense()
df = pd.DataFrame(doc_term_matrix, 
                  columns=count_vectorizer.get_feature_names(), 
                  index= ingredient_lists.keys())
```

```python
print(cosine_similarity(df, df))
```
